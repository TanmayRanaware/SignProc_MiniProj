```
X=[]
Y=[]
for i in range(0,100):
  start=np.random.randint(0,2517)
  end=start+100 
  x_sub_1=mel_db_1[:,start:end]
  X.append(x_sub_1)
  Y.append(0)
```

```
for k in range(0,100):
  start=np.random.randint(0,5157)
  end=start+100
  x_sub_3=mel_db_3[:,start:end]
  X.append(x_sub_3)
  Y.append(2)
```

```
for k in range(0,100):
  start=np.random.randint(0,5133)
  end=start+100
  x_sub_4=mel_db_4[:,start:end]
  X.append(x_sub_4)
  Y.append(3)
```

```
X=np.array(X)
Y=np.array(Y)
print(X.shape)
print(Y.shape)
```

```
X_train=np.reshape(X,(X.shape[0],X.shape[1],X.shape[2],1))
Y_train=np.reshape(Y,(Y.shape[0],1))
print(X_train.shape)
print(Y_train.shape)
```

```
x_train=X_train
y_train=Y_train
```

```
import tensorflow as tf
from tensorflow.keras import layers, models
def get_model():
  model = models.Sequential()
  model.add(layers.Conv2D(32, kernel_size = (3, 3), activation = 'relu', input_shape = [15,100,1]))
  model.add(layers.Conv2D(32, kernel_size = (3, 3), activation = 'relu'))
  model.add(layers.MaxPool2D(pool_size = (2, 2)))
  model.add(layers.BatchNormalization())
  model.add(layers.Conv2D(64, kernel_size = (3,3), activation = 'relu',strides=(2,2)))
  model.add(layers.MaxPool2D(pool_size = (2, 2)))
  model.add(layers.BatchNormalization())
  model.add(layers.Flatten())
  model.add(layers.Dense(128, activation = 'relu'))
  model.add(layers.Dense(4, activation = 'softmax'))
  model.summary()
  return model
```

```
classifier = get_model()
classifier.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```
x_test=[]
y_test=[]
for i in range(6):
  start=np.random.randint(0,2517)
  end=start+100 
  x_sub_1=mel_db_1[:,start:end]
  x_test.append(x_sub_1)
  y_test.append(0)

for j in range(6):
  start=np.random.randint(0,2257)
  end=start+100
  x_sub_2=mel_db_2[:,start:end]
  x_test.append(x_sub_2)
  y_test.append(1)

for k in range(6):
  start=np.random.randint(0,5157)
  end=start+100 
  x_sub_3=mel_db_3[:,start:end]
  x_test.append(x_sub_3)
  y_test.append(2)

for l in range(6):
  start=np.random.randint(0,5133)
  end=start+100 
  x_sub_4=mel_db_4[:,start:end]
  x_test.append(x_sub_4)
  y_test.append(3)
```

```
x_test=np.array(x_test)
y_test=np.array(y_test)
x_test=np.reshape(x_test,(x_test.shape[0],x_test.shape[1],x_test.shape[2],1))
y_test=np.reshape(y_test,(y_test.shape[0],1))
print(x_test.shape)
print(y_test.shape)
```

```
classifier.fit(x_train,y_train, validation_data=(x_test,y_test),epochs=30)
```

```
def prediction(index):
  print('Truth: {}'.format(lookup_table(y_test[index][0])))
  input = x_test[index]
  print('Image:')
  print(input.shape)
  input = input.reshape(1, input.shape[0], input.shape[1], input.shape[2])
  print(input.shape)
  pred = np.argmax(classifier.predict(input))
  print('Prediction: {}'.format(lookup_table(pred)))
```

```
classifier.evaluate(x_test,y_test)
```
